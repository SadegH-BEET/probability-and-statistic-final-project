<div dir = "rtl" style="direction:rtl;line-height:250%;">
	<font face="XB Zar" size=5>
اکنون نمودار همگرایی میانگین نمونه‌ها به امید ریاضی توزیع نمایی برحسب تعدادشان را برای $n=10000$ رسم‌ کنید.</br>
<b>راهنمایی:</b> بردار زیر را تشکیل دهید:
$$\left(
\frac{x_۱}{۱}, \frac{x_۱ + x_۲}{۲}, \frac{x_۱+x_۲+x_۳}{۳},
\frac{x_۱+x_۲+x_۳+x_۴}{۴}, \dots \right) .
$$
برای تشکل این بردار می‌توانید از تابع $\textrm{cumsum()}$ استفاده کنید.
</div>



```r
n <- 10000
lambda <- 0.1

samples <- rexp(n, rate = lambda)

sample_means <- cumsum(samples) / (1:n)

plot(1:n, sample_means, type = "l", xlab = "Index", ylab = "Partmeans")
abline(h = 1/lambda, col = "blue", lty = 2)

```

محمدپیام این بار سر تمرین فوتبال آمده است. او به طور متوالی به سمت دروازه شوت می‌زند. اولین شوت او گل می‌شود ولی شوت بعدی را گل نمی‌کند.
    از ضربه دوم به بعد به احتمال نسبت تعداد بارهایی که گل زده احتمال دارد که شوتش گل بشود.
    برای مثال اگر تا شوت ۳۰ ام موفق شده باشد ۱۰ شوت را گل کند، به احتمال
یک سوم شوت ۳۱ ام نیز گل خواهد شد.

الف) (۸۰ نمره)
    </font>
    </br>
صد ضربه اول محمدپیام را ۱۰۰۰۰ بار شبیه سازی کنید و
            نمودار احتمال گل شدن i توپ از این صد ضربه را بر حسب مقادیر مختلف i، رسم کنید.
            (۸۰ نمره)

```r
total_shots <- 100
num_simulations <- 100000

simulate_free_throws <- function(total_shots) {
  hits = 0
  misses = 0
  for (i in 1:total_shots)
  {
    if(i==1)
    {
      hits=hits+1
    }
    else if(i==2)
    {
      misses=misses+1
    }
    else
    {
    hit_probability = hits / (hits + misses)
    if (runif(1) < hit_probability)
    {
      hits = hits + 1
    }
    else
    {
      misses <- misses + 1
    }
    }
  }
  return(hits)
}


results <- numeric(num_simulations)
for (i in 1:num_simulations)
{
  results[i] = simulate_free_throws(total_shots)
}

count_hits <- table(results)/num_simulations
plot( count_hits, type = "l",  xlab = "Index", ylab = "Partmeans")

abline(h = 0.01, col = "blue", lty = 2)
```

<div dir = "rtl" style="direction:rtl;line-height:200%;">
<font face="XB Zar" color=#0000ff size=3>
میخواهیم نشان دهیم احتمال گل شدن j ضربه در n شوت برابر است با
$\frac{1}{n-1}$
با استقرا اثبات میکنیم:
<br>
پایه:برای$n=2$داریم
<br>
احتمال گل شدن ۱ ضربه برابر است با $\frac{1}{2-1}=1$
<br>
که برقرار است چون طبق صورت سوال ضربه اول ما گل شده و ضربه دوم گل نشده است
<br>
فرض استقرا:احتمال گل شده i ضربه از nضربه برابر است با:
$\frac{1}{n-1}$
<br>
حال احتمال گل شدن i ضربه در n+1ضربه را بررسسی می کنیم:
باید نشان دهیم این احتمال برابر است با:
$\frac{1}{n}$
<br>
دوحالت را در نظر میگیریم
<br>
فرض میکنیم ضربه آخر گل شده است که احتمال آن برابر است با
$\frac{i-1}{n}$
این ضربه را کنار میگذاریم  و طبق فرض استقرا به احتمال
$\frac{1}{n-1}$
i-1شوت از nشوت گل میشود
<br>
حال در نظر میگیریم که شوت اخر گل نشده است که احتمال آن برابر است با
$1-\frac{i}{n}$
<br>
حال بایدiضربه از n ضربه گل شوند که احتمال آن برابر است با
<br>
$\frac{1}{n-1}$
<br>
پس در کل احتمال گل شدن iضربه از n+1ضربه برابر است با:
<br>
$\frac{i-1}{n}\frac{1}{n-1} +( 1-\frac{i}{n})\frac{1}{n-1}=\frac{1}{n}$

```r
#DO NOT CHANGE ANYTHING IN THIS CELL.

test_simulate_step <- function() {
  # Test 1: One white square surrounded by black squares
  squares <- c(0, 1, 0, 0, 0)
  new_squares <- simulate_step(squares)
  # Expected result: One of the neighboring squares should become white
  if (any(new_squares == 0)) {
    print("Test 1 passed: One neighboring square turned white.")
  } else {
    print("Test 1 failed: No neighboring square turned white.")
  }

  # Test 2: Two white squares with one neighboring each
  squares <- c(0, 1, 0, 1, 0)
  new_squares <- simulate_step(squares)
  # Expected result: One of the neighboring squares should become white
  if (any(new_squares == 0)) {
    print("Test 2 passed: One neighboring square turned white.")
  } else {
    print("Test 2 failed: No neighboring square turned white.")
  }
}

test_check_adjacent_whites <- function() {
  # Test case 1: No white squares
  squares_1 <- c(1, 1, 1, 1, 1)
  if (check_adjacent_whites(squares_1)) {
    print("Test case 1 failed: No white squares")
  } else {
    print("Test case 1 passed")
  }

  # Test case 2: Single white square
  squares_2 <- c(1, 0, 1, 1, 1)
  if (check_adjacent_whites(squares_2)) {
    print("Test case 2 failed: Single white square")
  } else {
    print("Test case 2 passed")
  }

  # Test case 3: Adjacent white squares
  squares_3 <- c(1, 0, 0, 1, 1)
  if (check_adjacent_whites(squares_3)) {
    print("Test case 3 passed")
  } else {
    print("Test case 3 failed: Adjacent white squares")
  }

  # Test case 4: Non-adjacent white squares
  squares_4 <- c(1, 0, 1, 0, 1)
  if (check_adjacent_whites(squares_4)) {
    print("Test case 4 failed: Non-adjacent white squares")
  } else {
    print("Test case 4 passed")
  }
}



test_simulate_process <- function() {
  # Test 1: One white square surrounded by black squares
  N <- 3
  result <- simulate_process(N)
  # Expected result: 1/3, as one out of three squares remains white
  if (result == 1/3) {
    print("Test 1 passed: w(N)/N is correct.")
  } else {
    print("Test 1 failed: Incorrect w(N)/N value.")
  }

  # Test 2: Two white squares with one neighboring each
  N <- 4
  result <- simulate_process(N)
  # Expected result: 2/4, as two out of four squares remain white
  if (result == 2/4) {
    print("Test 2 passed: w(N)/N is correct.")
  } else {
    print("Test 2 failed: Incorrect w(N)/N value.")
  }
}

test_simulate_N_range <- function() {
  # Test: Simulation for N = 10, 20, 30 with 2 simulations each
  start_N <- 10
  end_N <- 30
  step_size <- 10
  num_simulations <- 2
  results <- simulate_N_range(start_N, end_N, step_size, num_simulations)
  # Expected result: Dataframe with 6 rows and 2 columns (N, w_N)
  if (nrow(results) == 6 && ncol(results) == 2) {
    print("Test passed: Correct number of rows and columns in results dataframe.")
  } else {
    print("Test failed: Unexpected number of rows or columns in results dataframe.")
  }
}
```

```r
# Function to simulate one step of the process
simulate_step <- function(squares) {
    indices<-which(squares==0)
    sampled_indice <- sample(indices,1)
    temp <- runif(1, 0, 1)
    chosen<-0
    if(sampled_indice==1)
    {
      chosen <-2
    }
    else if(sampled_indice == length(squares))
    {
      chosen = length(squares)-1
    }
    else if(temp <= 0.5)
    {
      chosen <- (sampled_indice - 1)
    }
    else
    {
      chosen <- (sampled_indice + 1)
    }
    if(squares[chosen] == 0)
    {
      squares[chosen]=1
    }
    return(squares)
}
```

```r
check_adjacent_whites <- function(squares) {
  # Identify the indices of white squares in the 'squares' vector.
  white_indices <- which(squares == 0)

  # If there are fewer than two white squares, return FALSE.
  if (length(white_indices) < 2) {
    return(FALSE)
  }

  # Check if there are adjacent white squares.
  for (i in seq_along(white_indices)[-length(white_indices)]) {
    if (white_indices[i + 1] - white_indices[i] == 1) {
      return(TRUE)
    }
  }

  # Return FALSE if there are no adjacent white squares.
  return(FALSE)
}
```

```r
test_check_adjacent_whites()
```

```r
simulate_process <- function(N) {
 squares <- c(1, rep(0, N), 1)

  while(check_adjacent_whites(squares))
    squares<-simulate_step(squares)

  remainingWhiteIndices<-which(squares==0)
  w<- length(remainingWhiteIndices)

  return (w/N)
}
```

```r
test_simulate_process()
```

```r
simulate_N_range <- function(start_N, end_N, step_size, num_simulations) {
  results <- data.frame(
    nValue = numeric(),
    rate = numeric()
  )
  numberOfRepetition <- seq(start_N, end_N, step_size)
  for (j in numberOfRepetition){
    for(i in 1:num_simulations){
      results <- rbind(results, data.frame(nValue = j, rate = simulate_process(j)))
    }
  }
  return (results)
}
```

```r
test_simulate_N_range()
```

```r
plot_results <- function(results) {
  library(ggplot2)
  p<-ggplot(results,aes(nValue,rate))+
    geom_point()+
    ggtitle("Simulation Results: w(N)/N vs. N")+xlab("N")+ylab("w(N)/N")+
    geom_smooth(method = "loess",se=FALSE)+
    theme_minimal()
  return (p)
}
```

```r
main <- function(start_N, end_N, step_size, num_simulations) {
  # Save the true value of 1/e
  true_value <- 1 / exp(1)

  # Perform simulations for a range of N values
  results <- simulate_N_range(start_N, end_N, step_size, num_simulations)

  # Calculate the mean of the calculated values
  mean_calculated_value <- mean(results[["rate"]])

  # Print the calculated value, true value, and absolute error
  cat("Calculated Value: ", mean_calculated_value, "\n")
  cat("True Value (1/e): ", true_value, "\n")
  cat("Absolute Error: ", abs(mean_calculated_value - true_value), "\n")

  # Plot the results
  plot_results(results)
}
```

```r
main(10, 3000, 50, 5)
```
